```python
import pandas as pd
```

```python
data = ''
with open('search_result.txt', encoding='utf8') as f:
    data = f.read()
data
```

```python
# \| (\d)권 (\d)강 (\d)일차 \| (.*) \| 45주 성경통독" href="(.*)&amp;list
import re
pattern = re.compile(r'\| (\d)권 (\d)강 (\d+)일차 \| (.*) \| 45주 성경통독" href="(.*)&amp;list')
result = pattern.findall(data)
result
```

```python
list =[[x[0], x[1], x[2], x[3], 'https://www.youtube.com' + x[4]] for x in result]
list
```

```python
df = pd.DataFrame(list, columns=['volume', 'chapter', 'day', 'title', 'url'])
df['volume'] = df['volume'].astype(int)
df['chapter'] = df['chapter'].astype(int)
df['day'] = df['day'].astype(int)
df.drop_duplicates(['title'], keep='first', inplace=True)
df
```

```python
# add date columns
# iloc[42] : 2025-03-10
# according to 42th row, fill the date column
# 'day' column is weekday (1: mon)
import datetime
from datetime import timedelta

start_date = datetime.date(2025, 1, 20)
for i in range(len(df)):
    df.loc[i, 'date'] = start_date + timedelta(days=(i + int(i / 6)))
df
```

```python
df.to_csv('url_list.csv', index=False, encoding='utf-8-sig')
```
